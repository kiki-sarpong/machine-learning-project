```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from sklearn.model_selection import train_test_split
```

```python
df_reviews = pd.read_csv("datasets/amazon_fine_food_review.csv")
df_reviews.head(10)
```

```python
score = df_reviews["Score"]
summary = df_reviews["Summary"]
rvs = df_reviews["Text"]
print(rvs)
```

```python
#cleaning the data 
data = rvs.copy()      #create a copy of the data for changes

for i,sentence in enumerate(data):
    dt = re.findall('<br />',sentence)
    if len(dt)> 1:
        print("ORIGINAL SENTENCE",dt,i)
        print(sentence)
        print()
        
        print("CLEANED SENTENCE")
        cd = re.sub('<br />','',sentence)
        print(cd)
        print()
        
    if (i>20):
        break
```

```python
train, test = train_test_split(data,test_size=0.3)
```
